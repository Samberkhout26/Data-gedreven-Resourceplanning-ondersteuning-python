```python
from sqlalchemy import create_engine

engine = create_engine("postgresql://samberkhout@localhost:5434/rister_prod17")
```

```python
df = pd.read_sql(
    """
    SELECT table_schema, table_name
    FROM information_schema.tables
    WHERE table_type = 'BASE TABLE'
    AND table_schema NOT IN ('pg_catalog', 'information_schema')
    ORDER BY table_schema, table_name
""",
    engine,
)

print(df.to_string())
```

```python
def get_table(schema, table, columns=None):
    cols = ", ".join(f'"{c}"' for c in columns) if columns else "*"
    df = pd.read_sql(f'SELECT {cols} FROM "{schema}"."{table}"', engine)
    df = df.rename(columns={"Id": f"{table.lower()}_id"})
    df = df.rename(columns={"Name": f"{table.lower()}name"})
    return df
```

# filter

```python
df = df[df["OrderStatus"].isin([5, 7, 8, 9, 10])]
```

```python
df = df[~df["timeregistration_id"].isna()]
```

```python
df.info()
```

```python
grouped = df.groupby(["EmployeeId"])["ManHourDuration"]
```

```python
df["med_gem_tijd"] = grouped.expanding().mean().shift(1).reset_index(level=[0, 1], drop=True)
df["med_std_tijd"] = grouped.expanding().std().shift(1).reset_index(level=[0, 1], drop=True)
df["med_aantal_opdrachten"] = (
    grouped.expanding().count().shift(1).reset_index(level=[0, 1], drop=True)
)
```

```python
df["taak_gem"] = (
    df.groupby(["ServiceId_x"])["ManHourDuration"]
    .expanding()
    .mean()
    .shift(1)
    .reset_index(level=[0, 1], drop=True)
)
```

```python
ervaring = df.groupby(["EmployeeId", "ServiceId_x"]).cumcount()
df["med_ervaring_bewerking"] = ervaring
```

```python
df = df.sort_values("Datum")
```

```python
df["med_klant_bezoeken"] = df.groupby(["EmployeeId", "ClientId"]).cumcount()
```

```python
df["med_totaal_opdrachten"] = df.groupby(["EmployeeId"]).cumcount()
```

```python
df["med_klant_ratio"] = df["med_klant_bezoeken"] / (df["med_totaal_opdrachten"] + 1)
```

```python
df["med_klant_gem_tijd"] = (
    df.groupby(["EmployeeId", "ClientId"])["ManHourDuration"]
    .expanding()
    .mean()
    .shift(1)
    .reset_index(level=[0, 1, 2], drop=True)
)
```

```python
df["med_klant_snelheid"] = (
    (df["med_klant_gem_tijd"] / df["med_gem_tijd"].replace(0, np.nan)).fillna(1.0).clip(0.1, 5.0)
)
```

```python
df["med_bewerking_gem_tijd"] = (
    df.groupby(["EmployeeId", "ServiceId_x"])["ManHourDuration"]
    .expanding()
    .mean()
    .shift(1)
    .reset_index(level=[0, 1, 2], drop=True)
)
```

```python
df["med_bewerking_snelheid"] = (
    (df["med_bewerking_gem_tijd"] / df["med_gem_tijd"].replace(0, np.nan))
    .fillna(1.0)
    .clip(0.1, 5.0)
)
```

```python
df["Datum"] = pd.to_datetime(df["Datum"])
```

```python
df["dag_van_week"] = df["Datum"].dt.dayofweek
df["maand"] = df["Datum"].dt.month
df["week"] = df["Datum"].dt.isocalendar().week
```

```python
df["week_sin"] = np.sin(2 * np.pi * df["week"] / 52)
df["week_cos"] = np.cos(2 * np.pi * df["week"] / 52)
df["dag_sin"] = np.sin(2 * np.pi * df["dag_van_week"] / 7)
df["dag_cos"] = np.cos(2 * np.pi * df["dag_van_week"] / 7)
df["maand_sin"] = np.sin(2 * np.pi * df["maand"] / 12)
df["maand_cos"] = np.cos(2 * np.pi * df["maand"] / 12)
```

```python
top10 = (
    df.groupby("orderserviceline_id")["UnitId"]
    .nunique()
    .sort_values(ascending=False)
    .head(20)
    .index
)

df[df["orderserviceline_id"].isin(top10)].groupby("orderserviceline_id").agg(
    unique_units=("UnitId", "nunique"), unit_names=("unitsname", set)
).sort_values("unique_units", ascending=False)
```

```python
# Pivot alle units naar kolommen
pivot = df.groupby(["orderserviceline_id", "unitsname"])["Amount"].sum().unstack(fill_value=0)

pivot.columns = [f"hoeveelheid_{col.lower().replace(' ', '_')}" for col in pivot.columns]
pivot = pivot.reset_index()

# Houd alleen de Uur rijen uit de originele df
df_uur = df[df["unitsname"] == "Uur"].drop_duplicates("orderserviceline_id")

# Merge de pivot kolommen erbij
df = df_uur.merge(pivot, on="orderserviceline_id", how="inner")
```

```python
hoeveelheid_cols = [
    "hoeveelheid_baal",
    "hoeveelheid_big_bag",
    "hoeveelheid_dag",
    "hoeveelheid_hectare",
    "hoeveelheid_kubieke_meter",
    "hoeveelheid_meter",
    "hoeveelheid_milimeter",
    "hoeveelheid_stuk",
    "hoeveelheid_ton",
]

available_hoeveelheid_cols = [c for c in hoeveelheid_cols if c in df.columns]
missing_hoeveelheid_cols = [c for c in hoeveelheid_cols if c not in df.columns]

if missing_hoeveelheid_cols:
    print(f"Missing hoeveelheid columns treated as 0: {missing_hoeveelheid_cols}")

if available_hoeveelheid_cols:
    df["heeft_extra_eenheid"] = df[available_hoeveelheid_cols].fillna(0).sum(axis=1) > 0
else:
    df["heeft_extra_eenheid"] = False
```
