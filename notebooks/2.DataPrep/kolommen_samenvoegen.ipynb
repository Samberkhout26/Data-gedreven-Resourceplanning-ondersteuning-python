# Rister + Werkxpert -> gefilterd op verbinding
Laadt beide CSV's en houdt alleen de kolommen over die in de verbindingstabel staan.
Rister-kolommen worden hernoemd naar de Werkexpert-naam.

```python
import pandas as pd
import numpy as np

KOLOM_MAPPING = {
    'EmployeeId':               'URENVERANTW_MEDID',
    'ServiceLineId_x':          'DIENST_ART_ID',
    'PlanningGroupId':          'BEWERKING_ID',
    'ClientId':                 'RELATIE_ID',
    'servicelinesname':         'DIENST_ART_OMS',
    'equipmentname':            'MACH_OMS',
    'PostalCode':               'REL_POSTCODE',
    'lat':                      'lat',
    'lon':                      'lon',
    'Datum':                    'URENVERANTW_DATUM',
    'dag_sin':                  'dag_sin',
    'dag_cos':                  'dag_cos',
    'maand_sin':                'maand_sin',
    'maand_cos':                'maand_cos',
    'week_sin':                 'week_sin',
    'week_cos':                 'week_cos',
    'med_gem_tijd':             'med_gem_tijd',
    'med_std_tijd':             'med_std_tijd',
    'med_aantal_opdrachten':    'med_aantal_opdrachten',
    'med_ervaring_bewerking':   'med_ervaring_bewerking',
    'taak_gem':                 'taak_gem',
    'med_klant_bezoeken':       'med_klant_bezoeken',
    'med_klant_ratio':          'med_klant_ratio',
    'med_klant_snelheid':       'med_klant_snelheid',
    'med_bewerking_snelheid':   'med_bewerking_snelheid',
    'med_klant_gem_tijd':       'med_klant_gem_tijd',
    'med_bewerking_gem_tijd':   'med_bewerking_gem_tijd',
    'med_totaal_opdrachten':    'med_totaal_opdrachten',
    'hoeveelheid_big_bag':      'hoeveelheid_volume',
    'hoeveelheid_dag':          'hoeveelheid_gewicht',
    'hoeveelheid_hectare':      'hoeveelheid_stuks',
    'hoeveelheid_kubieke_meter':'hoeveelheid_aanwezig',
    'TenantId_x':               'con',
    'hoeveelheid_uur' :         'REAL_WORKED_TIME'
}

RISTER_ONLY = ['EquipmentGroupTypes', 'planninggroupsname', 'hoeveelheid_baal']
WERKXPERT_ONLY = []
```

```python
# -- Rister CSV inladen, filteren en hernoemen naar Werkexpert-namen ----------
df_rister = pd.read_csv('rister.csv', low_memory=False)

rister_cols_all = list(KOLOM_MAPPING.keys()) + RISTER_ONLY
beschikbaar_rister = [c for c in rister_cols_all if c in df_rister.columns]
ontbrekend_rister = [c for c in rister_cols_all if c not in df_rister.columns]

df_rister = df_rister[beschikbaar_rister]

# Mapping richting omdraaien voor rister: rister_kolom -> werkexpert_kolom
rename_map_rister = {k: v for k, v in KOLOM_MAPPING.items() if k in beschikbaar_rister}
df_rister = df_rister.rename(columns=rename_map_rister)

print(f'Rister shape:  {df_rister.shape}')
print(f'Ontbrekend:    {ontbrekend_rister}')
df_rister.head()
```

```python
# -- Werkxpert CSV inladen en filteren (namen blijven Werkexpert) --------------
df_werkxpert = pd.read_csv('werkexpert.csv', low_memory=False)

# Maak de gevraagde kolomnamen uniek (zonder volgorde te verliezen).
werkxpert_cols_all = list(dict.fromkeys(list(KOLOM_MAPPING.values()) + WERKXPERT_ONLY))
beschikbaar_wx = [c for c in werkxpert_cols_all if c in df_werkxpert.columns]
ontbrekend_wx = [c for c in werkxpert_cols_all if c not in df_werkxpert.columns]

df_werkxpert = df_werkxpert[beschikbaar_wx]

print(f'Werkxpert shape:  {df_werkxpert.shape}')
print(f'Ontbrekend:       {ontbrekend_wx}')
df_werkxpert.head()
```

```python
df_rister['bron']    = 'rister'
df_werkxpert['bron'] = 'werkxpert'

df = pd.concat([df_werkxpert, df_rister], ignore_index=True)

print(f'Gecombineerd shape: {df.shape}')
df.head()
```

```python
df.tail()
```

```python
# === Geschiktheidsscore berekenen ===
df['norm_ervaring_bewerking'] = df.groupby('con')['med_ervaring_bewerking'].transform(
    lambda x: x / x.max() if x.max() > 0 else 0
)
df['norm_klant_bezoeken'] = df.groupby('con')['med_klant_bezoeken'].transform(
    lambda x: x / x.max() if x.max() > 0 else 0
)
df['suitability_score'] = (
    0.6 * df['norm_ervaring_bewerking'] + 
    0.4 * df['norm_klant_bezoeken']
).clip(0, 1).astype(np.float32)

print(f"Geschiktheidsscore verdeling:")
print(df['suitability_score'].describe())
```

```python
# ── Opslaan ───────────────────────────────────────────────────────────────────
df.to_csv('dataframe_gecombineerd.csv', index=False)
print('Klaar: dataframe_gecombineerd.csv')
```
